```python
import glob
import xml.etree.ElementTree as et
import pandas as pd
import datetime
```

```python
import glob
path = '/Users/eadaoinburgess/Desktop/Data/YRDonegal'


for file in glob.glob(path + "/*YRDonegal"):
    with open(file, 'r', encoding="utf-8") as content:
        xtree = et.parse(content)
        xroot = xtree.getroot()
        data = []
        data1= []
        for i in xtree.iterfind(".//precipitation"): 
            data1.append(dict(i.items()))
        df2 = pd.DataFrame(data1)
        
        for node in xroot.iter("time"):
            data.append(dict(node.items()))
        df = pd.DataFrame(data)
            
        df2["from"]= df["from"]
        df2["to"]=df["to"]
        df2['from']=pd.to_datetime(df2['from'])
        df2['new_date'] = [d.date() for d in df2['from']]
        
        
        current_date = df2["new_date"][0]

        oneday= current_date + datetime.timedelta(days=1)
        threeday = current_date + datetime.timedelta(days=3)
        fiveday= current_date + datetime.timedelta(days=5)
    
        df2["value"] = pd.to_numeric(df2["value"])
        five = df2.loc[df2['new_date']== fiveday, 'value'].sum()
        three = df2.loc[df2['new_date']== threeday, 'value'].sum()
        one = df2.loc[df2['new_date']== oneday, 'value'].sum()
        
        
        data2 = {'Provider': ['YR'],'Location': ['Donegal'],'One_Day_Ahead':[one],'DateOne':[oneday],'Three_Days_Ahead':[three],'DateThree': [threeday] ,'Five_Days_Ahead':[five],'DateFive': [fiveday], 'Date_Scraped': [current_date]} 

        df1 = pd.DataFrame(data2)
        
        import os.path as paths
        if paths.exists("forecastYR.csv"):
            forecastData = pd.read_csv("forecastYR.csv") 
            forecastData = forecastData.append(df1)
            forecastData.to_csv("forecastYR.csv",index=False)
        else:
            df1.to_csv("forecastYR.csv",index=False)
        
    
```
